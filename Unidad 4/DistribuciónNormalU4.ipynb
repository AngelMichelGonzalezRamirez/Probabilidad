```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Generar datos para la distribución normal.

media = 70     #Media
desv_std = 4   #Desviación estándar
num_datos = 500

#Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Muestra los primeros 50
```

```python
# Crear un rango de valores para crear una densidad

x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

  #Calcular la PDF de la distribución normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std **2))
```

```python
#Para no imprimir toda la lista de valores (500), desplegamos los primeros 50

print ("Imprimir los primeros 50 valores de X", x[0:50], " , y los primeros 50 pdf", pdf[0:50])
```

```python
#Graficar histograma
#Generamos un histograma de las edades para visualizar cuales son los grupos de edades más comunes

plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
#Visualizar la PDF

plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Pesos en Kg. de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Graficar Histograma
#Aquí estamos uniendo el histograma y la gráfica de función de densidad y probabilidad.

plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha= 0.4, density= True, color= 'b')
plt.plot(x, pdf, 'k', linewidth=2 )
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

#Ejercicio 2

```python
plt.figure(figsize=(6,4))

#Parametros de la disitribución normal
media = 70       #Media
desv_std = 4     #Desviación estándar
total_estudiantes = 500

#Variable para cálculo de la probabilidad de que un estudiante que pese menos de 75 kg
peso_limite= 79

#Primero necesitamos calcular la probabilidad de que un estudiante pese menos
#de 75 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal.

prob_menos_75kg= norm.cdf(peso_limite,media, desv_std)
```

```python
#Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Npumero de estudiantes que pesan menos de 75 kg: ", estudiantes_menos_75kg)

#Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()

```

####Cuantos estudiantes pesan mas de 79 kg.

```python
plt.figure(figsize=(6, 4))

#Parametro de la distribución normal
media = 70      #Media
desv_std = 4    #Desviación estándar
peso_limite = 79    #Peso Limite
total_estudiantes=500

#Calculamos la probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite,media, desv_std)

#La probabilidad de que un estudiante pese más de 79 kg es complementaria
prob_mas_79kg = 1 - prob_menos_79kg

#Calculamos el número aproximado de estudiantes que pesan más de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("Probabilidad de que un estudiante pese más de 79 kg:", prob_mas_79kg)
print("Número aproximado de estudiantes que pesan más de 79 kg:", estudiantes_mas_79kg)
```

```python
#Calcular la función de densidad de probabilidad de la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x >= peso_limite), color='skyblue', alpha= 0.5)
plt.title('Probabilidad de que un estudiante pese más de 79 kg, \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

### Problema 1 de distribución binomial en Python/Trabajo de plataforma:

```python
plt.figure(figsize=(6, 4))

#Parametro de la distribución normal
media = 70      #Media
desv_std = 4    #Desviación estándar
total_estudiantes=500

# Rangos de peso utilizados en el ejercicio:
peso_min = 68
peso_max = 72

# Calculamos la probabilidad de que un estudiante pese entre 68 y 72 kg:
prob_entre_68_y_72kg = norm.cdf(peso_max, media, desv_std) - norm.cdf(peso_min, media, desv_std)

#Al restar estas dos probabilidades acumuladas, obtenemos la probabilidad de
#que un estudiante esté en el rango de peso entre peso_min y peso_max.


# Calculamos el número de estudiantes que pesan entre 68 y 72 kg:
estudiantes_entre_68_y_72kg = total_estudiantes * prob_entre_68_y_72kg

print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", prob_entre_68_y_72kg)
print("Número de estudiantes que pesan entre 68 y 72 kg:", estudiantes_entre_68_y_72kg)
```
