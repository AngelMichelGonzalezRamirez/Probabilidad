```python
import math as mt #Para utlizar factorial.
from itertools import permutations
from itertools import product
import numpy as np
```

```python
asistentes=["A1","A2","A3","A4","A5","A6","A7","A8","A9","A10"]
n=len(asistentes)
r=4 #¿cómo agrupar?

Pr=mt.factorial(n) / (mt.factorial(n-r))

print ("Existen" , Pr, " posibles permutaciones sin repetición del total de",
       n, "asistentes para calificar cada pregunta ", "en grupos de ",r,"en ",r)

print ("Aquí las permutaciones")
permutaciones_asistentes=list(permutations(asistentes,r))

#Imprimimos las permutaciones:
for permutacion in permutaciones_asistentes:
  print (permutacion)
```
