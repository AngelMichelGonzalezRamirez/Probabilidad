##Ejemplo 3: Histogramas de dos grupos.
1. En este ejemplo generamos 100 numeros aleatorios reales que representa la producción diaria de 100 telares (medidas en metros).
2. Generar otra lista con la producción de 100 telares para una segunda fábrica.
3. Con estas lista de valores se pretende determinar las medidas de tendencia central, es decir, la media, la moda, la mediana, la desviación estandar.
4. Graficar un histograma para cada fábrica.



```python
#Librería para generar numeros aleatorios.
import random
```

```python
fabrica1= []
for i in range(100):
  numaleatorio= round(random.uniform (50,100),2)  #Round me redondeo los numeros a dos decimales, por eso el ,2.
  fabrica1.append(numaleatorio)    #Con el método append, puedes agregar un elemento al final de una lista. Es como colocar un objeto al final de una fila.
print(fabrica1)
```

```python
fabrica2= []
for i in range(100):
  numaleatorio2= round(random.uniform (50,100),2)
  fabrica2.append(numaleatorio2)
print(fabrica2)
```

```python
#Librería de Python para graficar valores.
import matplotlib.pyplot as plt

#"as plt", es el alias que le dimos debido a que el nombre de la libreria es muy extenso
```

```python
plt.hist(fabrica1)  #.hist de histograma, o mejor dicho gráfica.
plt.title("Producción diaria de la Fábrica 1")
```

```python
plt.hist(fabrica2)  #.hist de histograma
plt.title("Producción diaria de la Fábrica 2")
```

```python
media=mean(fabrica1)
mediana=median(fabrica1)
moda=mode(fabrica1)
print("Valores de la Fábrica 1")
print("La media es igual a",media)
print("La mediana es igual a",mediana)
print("La moda es igual a",moda)
```

```python
media=mean(fabrica2)
mediana=median(fabrica2)
moda=mode(fabrica2)
print("Valores de la Fábrica 2")
print("La media es igual a",media)
print("La mediana es igual a",mediana)
print("La moda es igual a",moda)
```

```python
#Generar dos subplots de 1 renglón por dos columnas para desplegar los dos histogramas, una al lado de la otra.
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2 , figsize=(10,5)) #subplot sirve para poner el area de trazado de nuestros gráficos.
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

##Tablas de Frecuencia

```python
#Generamos una lista vacía que representa la edad de un grupo de personas.
edades=[]
for i in range(100):
  numero_aleatorio=random.randint (18,70)
  edades.append(numero_aleatorio)
print (edades)
```

```python
import pandas as pd
#Pandas es una biblioteca de Python que se utiliza principalmente para análisis de datos y manipulación de datos estructurados.
```

```python
df_edades = pd.DataFrame(edades,columns=["Edad"])
df_edades #Imprimo el dataframe.
```

```python
#Calcular la tabla de distribución de frecuencias.
tabla_frecuencias=df_edades["Edad"].value_counts().reset_index() #Para ordenarlo por edades.
tabla_frecuencias.columns = ["Edad","Frecuencia"] #Agrego columna de Frecuencia.
tabla_frecuencias = tabla_frecuencias.sort_values(by="Edad") #Ordena los valores por Edad.
tabla_frecuencias
```

```python
tabla_frecuencias["Frec. Relativa"] = tabla_frecuencias["Frecuencia"] / 100
```

```python
tabla_frecuencias
```

```python
#Calcular la Frecuencia Porcentual:
tabla_frecuencias["Frec. Porcentual"] = tabla_frecuencias["Frecuencia"]/ 100*100
tabla_frecuencias
```

```python
#Calcular la frecuencia acumulada:
#Agregamos una nueva columna con el cálculo
tabla_frecuencias["Frec. Acumulada"] = tabla_frecuencias["Frecuencia"].cumsum()
tabla_frecuencias
```

##Regla de Sturges
1.- Es una regla que sirve para calcular el número de clases o intervalos idóneo, en los que se debe de dividir un conjunto de datos.

$
c = 1+Log_2(N)
$

```python
#Librería de funciones matemáticas:
import math
```

```python
print(edades)
```

```python
plt.hist(edades,alpha=0.5,rwidth=0.90)
plt.title("Histograma de edades basado en valores estandar de matplotlib")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
```

```python
df_edades_sturges = pd.DataFrame({"Edad":edades})
df_edades_sturges
```

```python
#Calcular el número de intervalos o clases utilizando la regla de Sturges:
n = len(edades)
k = int(1+math.log2(n))
print("Valor de n=",n,"\n K o número de intervalos o clases=",k)
```

```python
tabla_frec_sturges = pd.cut(df_edades_sturges["Edad"], bins = k).value_counts().reset_index()
tabla_frec_sturges
```

```python
tabla_frec_sturges.columns=["Clase","Frecuencia"]
tabla_frec_sturges
```

```python
#Ordenar la tabla de frecuencias por la clase o el intervalo:
tabla_frec_sturges = tabla_frec_sturges.sort_values(by="Clase")
tabla_frec_sturges
```

```python
tabla_frec_sturges["Frec.Relativa"] = tabla_frec_sturges["Frecuencia"]/n
tabla_frec_sturges["Frec.Porcentual"] = tabla_frec_sturges["Frecuencia"]/n*100
tabla_frec_sturges["Frec.Acumulada"] = tabla_frec_sturges["Frecuencia"].cumsum()
tabla_frec_sturges
```

```python
plt.hist(df_edades_sturges["Edad"], bins=k, alpha=0.5, rwidth=.95)
plt.title("Histograma de edades basado en la regla de Sturges")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
```
